```python
import pandas as pd

df = pd.read_parquet("../anbima_data.parquet")
df
```

```python
min_date = df["ReferenceDate"].min()
df_new = pd.read_parquet("...")[df.columns]
df = (
    pd.concat([df, df_new])[df.columns]
    .drop_duplicates()
    .sort_values(["ReferenceDate", "BondType"])
    .query("ReferenceDate >= @min_date")
    .reset_index(drop=True)
)
df
```

```python
df.info()
```

```python
rdate = pd.to_datetime("03-01-2024", dayfirst=True)
df.query("ReferenceDate == @rdate").reset_index(drop=True)
```

```python
df.to_parquet("../anbima_data.parquet", index=False, compression="gzip")
```
