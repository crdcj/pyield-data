```python
byears = df["BDaysToExp"] / 252
dv01_price = 100_000 / ((1 + df["SettlementRate"] + 0.0001) ** byears)
df["DV01"] = (df["SettlementPrice"] - dv01_price).astype("Float64")
df
```
